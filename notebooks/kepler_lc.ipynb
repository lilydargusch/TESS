```python
from astropy.table import Table
import matplotlib.pyplot as plt
```

```python
lc = Table.read('http://archive.stsci.edu/hlsps/kepler-smear/hlsp_kepler-smear_kepler_phot_kplr8766240_kepler_v1.0_lc.fits')
```

```python
BJD = lc['BJD']
FLUX = lc['FLUX_CORR_4']
FLUX2 = lc['FLUX_CORR_8']
FLUX3 = lc['FLUX']
```

```python
plt.plot(BJD, FLUX, color='black', markersize=1, label='4 CBV')
plt.plot(BJD, FLUX2, color='green', markersize=1, label='8 CBV')
plt.plot(BJD, FLUX3, color='red', markersize=1, label='Raw Flux')
plt.xlabel('Time (BJD Days)')
plt.ylabel('Flux')
plt.title('HD186994')
plt.legend()
```
