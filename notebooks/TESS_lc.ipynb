```python
import lightkurve as lk
import matplotlib.pyplot as plt
```

```python
name = 'HD_185351'
lc = lk.search_lightcurve(name, mission='TESS')
lc
```

```python
lc = lk.search_lightcurve(name, mission='TESS').download_all()
```

```python
lc.plot(normalize=True)
plt.title(name)
plt.legend(["SPOC (14)", "TESS_SPOC (14)", "QLP (14)","SPOC (15)", "TESS_SPOC (15)", "QLP (15)"], bbox_to_anchor=(1, 0.5))
```
